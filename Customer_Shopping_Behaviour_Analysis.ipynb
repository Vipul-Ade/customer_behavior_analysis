```python
import pandas as pd

df = pd.read_csv('/content/drive/My Drive/Practise_files/customer_shopping_behavior.csv')

df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns = {'purchase_amount_(usd)' : 'purchase_amount'})
```

```python
df.columns
```

```python
# Create new column Age group

labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['Age_group'] = pd.qcut(df['age'], q = 4, labels=labels)
```

```python
df[['age', 'Age_group']]
```

```python
# create column purchase frquency days

frequency_mapping = {
    "Fortnightly" : 14,
    "Weekly" : 7,
    "Monthly" : 30,
    "Quarterly" : 90,
    "Bi-Weekly" : 14,
    "Annually" : 365,
    "Every 3 Months" : 90
}

df['Purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['frequency_of_purchases', 'Purchase_frequency_days']].drop_duplicates() #see all unique pairs
```
